# Operaciones b谩sicas con matrices

NumPy nos permite realizar varias operaciones con matrices, como suma, resta, producto escalar, producto de matrices y divisi贸n. Veamos c贸mo hacerlo:

```python
# Crear dos matrices para las operaciones
A = np.array([[1, 2], [3, 4]])
B = np.array([[5, 6], [7, 8]])

# Suma de matrices
print("A + B =\n", A + B)

# Resta de matrices
print("\nA - B =\n", A - B)

# Producto escalar
print("\n2 * A =\n", 2 * A)

# Producto de matrices
print("\nA @ B =\n", A @ B)

# Divisi贸n de matrices
print("\nA / B =\n", A / B)
```

# Algunas operaciones especiales con matrices

Dada la matriz $ A = \begin{bmatrix} 1 & 0 & 1 \\ 4 & -1 & 4 \\ 5 & 6 & 7 \end{bmatrix} $, encuentra la traza de la matriz inversa de $ A $.

```python
import numpy as np

# Se define la matriz A
A = np.array([[1, 0, 1],
              [4, -1, 4],
              [5, 6, 7]])

# Se calcula la inversa de A
A_inv = np.linalg.inv(A)

# Se calcula la traza de la matriz inversa
traza = np.trace(A_inv)

print("Traza de la inversa de A:", traza)
```

### Objetivos del problema
Crear el Tablero de Juego:

Implementa una funci贸n para crear un tablero de juego de dimensiones n x m, donde cada celda puede estar viva (1) o muerta (0). Inicializa el tablero con un patr贸n inicial.

### Definir las reglas del juego:

Cada celda en el tablero tiene 8 vecinos. Las reglas para la evoluci贸n del estado de las celdas son:
Una celda viva con menos de dos celdas vecinas vivas muere por subpoblaci贸n.
Una celda viva con dos o tres celdas vecinas vivas sigue viva en la siguiente generaci贸n.
Una celda viva con m谩s de tres celdas vecinas vivas muere por sobrepoblaci贸n.
Una celda muerta con exactamente tres celdas vecinas vivas se convierte en una celda viva por reproducci贸n.

### Simular la evoluci贸n:

Implementa una funci贸n para actualizar el tablero siguiendo las reglas del juego. Esta funci贸n deber铆a generar la nueva configuraci贸n del tablero despu茅s de un n煤mero espec铆fico de iteraciones.

### Visualizar la simulaci贸n:

Utiliza Matplotlib para visualizar la evoluci贸n del tablero a lo largo de las iteraciones. Muestra cada estado del tablero como una imagen en una animaci贸n.

```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Se crea tablero inicial
def crear_tablero(n, m):
    return np.random.choice([0, 1], size=(n, m))

# ConSe cuenta vecinos vivos
def contar_vecinos(tablero, x, y):
    n, m = tablero.shape
    vecinos = 0
    for i in range(max(0, x-1), min(x+2, n)):
        for j in range(max(0, y-1), min(y+2, m)):
            if (i, j) != (x, y):
                vecinos += tablero[i, j]
    return vecinos

# Se aplican reglas del juego para actualizar el tablero
def actualizar_tablero(tablero):
    n, m = tablero.shape
    nuevo = np.copy(tablero)
    for i in range(n):
        for j in range(m):
            vivos = contar_vecinos(tablero, i, j)
            if tablero[i, j] == 1:
                if vivos < 2 or vivos > 3:
                    nuevo[i, j] = 0
            else:
                if vivos == 3:
                    nuevo[i, j] = 1
    return nuevo

# Simulaci贸n y animaci贸n del juego
def simular_vida(n, m, generaciones):
    tablero = crear_tablero(n, m)
    fig, ax = plt.subplots()
    img = ax.imshow(tablero, cmap='binary')
    ax.set_title("Juego de la Vida - Conway")
    ax.axis('off')

    def actualizar(frame):
        nonlocal tablero
        tablero = actualizar_tablero(tablero)
        img.set_data(tablero)
        return [img]

    anim = animation.FuncAnimation(fig, actualizar, frames=generaciones, interval=300, blit=True)
    plt.show()

# Se ejecuta la simulaci贸n
simular_vida(20, 20, 50)  

```
